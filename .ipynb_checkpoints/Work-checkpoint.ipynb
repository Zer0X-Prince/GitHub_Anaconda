### String

```python
# f string and type
x = "Hello World"
y = "And Me"
print(f"Hello {x}".title())
print(f"Hello {x}".upper())
print(f"Hello {x}".lower())
print(f"Hello {x}".count("l"))
print(f"Hello {x}".replace("l","Q"))
print(f"Test {y} {x}")
print(type(x))
```

### List 

```python
countries = ['USA', 'India', 'UK']
print(countries[1])
print(countries[-2])
print(countries[0:3:1])
print(countries[:3])
print(countries[0:])
countries.append('XYZ')
countries.extend('X')
countries.insert(1, 'Z')
print(countries)
```

```python
countries2 = ['A', 'B', 'C']
```
