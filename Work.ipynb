```python
# add multiple columns using assign()
score1 = np.random.uniform(100, 1000, size = 26)
score2 = np.random.uniform(100, 1000, size = 26)
```

```python
series1 = pd.Series(score1, index = np.arange(0,26))
series2 = pd.Series(score2, index = np.arange(0,26))
df_mc = df_mc.assign(Score1 = series1, Score2 = series2)
df_mc.head()
```
